```python
def plot_loss_accuracy(nn_, file_name):
    """
    generate a plot showing loss and accuracy among train and valid data
    ----------
    nn_ : instance of Neural Network class
    """
    # get structural info
    lr = f"learning rate: {nn_.learning_rate}"
    bias = "bias: True" if nn_.with_bias else "bias: False"
    keep_p = "keep_p: " + ",".join([str(layer.keep_p) for layer in nn_.layers if layer.keep_p is not None])
    nodes = f"no of nodes: {','.join([str(layer.unit_num) for layer in nn_.layers])}" 
    if nn_.f_cost == squared_err:
        f_cost_name = "Mean Squared Error"
    elif nn_.f_cost == cross_entropy:
        f_cost_name = "Cross-entropy"
    else:
        f_cost_name = ""
    params = [lr, bias, keep_p, nodes]
    # plot
    figure, ax1 = plt.subplots()
    plt.suptitle("Accuracy and loss over epochs")
    x = [i+1 for i in range(len(nn_.train_accuracies))]
    ax1.plot(x, nn_.train_accuracies, color="teal", label="train acc")
    ax1.plot(x, nn_.test_accuracies, color="skyblue", label="valid acc")
    ax2 = ax1.twinx()
    ax2.plot(x, nn_.train_losses, color="darkred", label="train loss", alpha=0.7)
    ax2.plot(x, nn_.test_losses, color="tomato", label="valid loss", alpha=0.7)
    ax1.set_ylabel("Accuracy")
    ax1.set_title("Params: {}".format(",".join(params)), fontsize=10)
    ax2.set_ylabel(f"{f_cost_name}")
    ax1.set_xlabel("Epochs")
    h1, l1 = ax1.get_legend_handles_labels() # handle, label
    h2, l2 = ax2.get_legend_handles_labels()
    ax1.legend(h1+h2, l1+l2, loc='lower right')
    
    # output
    df_ = pd.DataFrame({"train_acc":nn.train_accuracies,
                        "test_acc": nn.test_accuracies,
                        "train_loss":nn.train_losses, 
                        "test_loss": nn.test_losses})
    # GoogleColab
    #     df_.to_csv(f"/content/drive/My Drive/Colab Notebooks/INM702/output/{file_name}.csv")
    #     plt.savefig(f"/content/drive/My Drive/Colab Notebooks/INM702/output/{file_name}.png", bbox_inches="tight")
    df_.to_csv(f"./output/{file_name}.csv")
    plt.savefig(f"./output/{file_name}.png", bbox_inches="tight")
    min_loss = df_["test_loss"].min()
    max_acc = df_["test_acc"].max()
    print(f"min_loss:{min_loss}, max_acc:{max_acc}")
    
    plt.show()
```

# Implementation of Neural Network
In this section, the neural network is developed from scratch.

## Global variable

```python
from enum import Enum
class LType(Enum):
    """
    This represents a type of layer in a neural network
        INP: input layer
        HID: hidden layer
        OUT: output layer
    """
    INP = 1
    HID = 2
    OUT = 3
```

## Cost function

```python
import warnings
def cross_entropy(pred, target):
    """
    return numpy.float
        cross entropy. only called from the output layer.
    ----------
    pred : numpy.array
        predicted output generated by softmax
    labels: numpy.array
        true labels of probability
    """
    delta = 1e-7 # to avoid zero divide error in np.log()
    with warnings.catch_warnings():
        warnings.filterwarnings("error")
        try:
            res = np.mean(-np.sum(target * np.log(pred + delta), axis=1, keepdims=True))
        except RuntimeWarning as e:
            print(f"RuntimeWarning in calculating cross entropy: {e}")

    return res
```

```python
def squared_err(pred, labels):
    """
    return numpy.float
        squared error
    ----------
    pred : numpy.array
        predicted output
    labels: numpy.array
        true labels
    """
    return np.mean(np.sum((1/2) * (pred - labels) ** 2, axis=1, keepdims=True))
```

## Activation function

```python
def sigmoid(wx):
    """
    return numpy.float
        it returns the values passed through sigmoid function
    ----------
    wx : numpy.array
        input for the activation function
    """
    return 1 / (1 + np.e ** -wx)

def relu(wx):
    """
    return numpy.float
        it returns the values passed through ReLU function
    ----------
    wx : numpy.array
        input for the activation function
    """
    return np.maximum(0.0, wx)

def softmax(wx):
    """
    return numpy.float
        it returns the values passed through Softmax function
    ----------
    wx : numpy.array
        input for the activation function
    """
    wx -= np.max(wx, axis=1, keepdims=True)
    sf = np.exp(wx) / np.sum(np.exp(wx), axis=1, keepdims=True)
    return sf
```

## Derivative of Cost Function

```python
def differentiate_cost_f(f_cost, f_act, tgt, op):
    """
    return numpy.array
        If f_cost is NOT cross-entropy, return a differentiated cost function 
        If f_cost is cross-entropy, return a differentiated cost and activation functions
    ----------
    f_cost : Layer.f_act
        any cost functions such as squared_error and cross_entropy
    f_act: function
        any activation functions such as relu, sigmoid and softmax
    tgt: numpy.array
        true labels
    op: numpy.array
        output of layer, output of activation function
    """
    if f_cost == squared_err:
        return (tgt - op) / len(tgt)
    elif f_cost == cross_entropy:
        assert f_act == softmax, "Please use softmax in the output layer"
        return (tgt - op) / len(tgt)
    else:
        print(f"Unexpected cost function: {f_cost}", file=sys.stderr)
        sys.exit(1)
```

## Derivative of activate functions

```python
def differentiate_act_f(f_, output_):
    """
    returning a result from a differentiated activation function
    ----------
    f_ : Layer.f_act
        activation function in Layer class
    output_: numpy.array
        it is an output of activation function
    """
    if f_ == sigmoid:
        return output_ * (1 - output_)
    elif f_ == relu:
        pd_relu = np.vectorize(lambda op: 0 if op <= 0 else 1)
        return pd_relu(output_)
    elif f_ == softmax:
        return 1 # pd_softmax / pd_wx is already considered in function(differentiate_cost_f)
```

## Layer class

```python
class Layer:
    """
    Representing a single layer from input, hidden or output layer.
    Constructor is only called from NeuralNetwork
    """
    def __init__(self, nn_, structure, ealier_unit_num, with_bias, fix_seed):
        """
        nn: object of NeuralNetwork class
        structures: dict
            type: Enum of LType, unit_num: the number of units, f_act: activation function
        ealier_unit_num: int
            number of units in an ealier layer
        with_bias: boolean
            true if introducing bias in the layer
        fix_seed: boolean
            true if removing randomness; for debugging purpose
        """
        self.type = structure["type"] # type of layer (imput, hidden, output)
        self.with_bias = with_bias
        self.fix_seed = fix_seed
        self.unit_num = structure["unit_num"] + 1 if self.with_bias else structure["unit_num"] # number of neurons in this layer
        self.f_act = structure["f_act"] if "f_act" in structure else relu # activation function
        self.x = None # input of this layer
        self.diff_w = None # gradient before multiplying learning rate
        self.prev_diff_w = None if nn_.momentum is None else 0 # gradient of the previous iteration
        self.wx = None # input of the activation function
        self.output = None # output of the layer
        self.delta = None # temporal computational result while backpropagation
        self.keep_p = structure["keep_p"] if "keep_p" in structure else None # probability of activated units; 0 to 1;
        self.inverted_drop_mask = None # for masking output of the layer
        self.w_hist = [] # history of weight per epoch for visualization
        self.w = None if structure["type"] == LType.INP else self.__get_init_w(ealier_unit_num) # weight of this layer
        
    def __get_init_w(self, ealier_unit_num):
        """
        return the initialized weights
        ----------
        ealier_unit_num: int
            the number of units in the earlier layer
        """
        # only accept LType.HID, LType.OUT
        if self.with_bias:
            shape = (self.unit_num - 1, ealier_unit_num)
        else:
            shape = (self.unit_num, ealier_unit_num)
        sd = 1 / np.sqrt(shape[1]) # According to Xavier, this SD is good for the weight initialisation
        sd *= np.sqrt(2) if self.f_act == relu else 1 # According to He, it is good initialised values for ReLU
        # truncate if a weight is more than 2 * SD, it is the same implementation with TensorFlow
        # range is calculated by a, b = (myclip_a - my_mean) / my_std, (myclip_b - my_mean) / my_std, which is 2 in this case
        range_ = 2 
        if self.fix_seed: 
            np.random.seed(seed=1) # debug
        init_w = truncnorm(-range_, range_, loc=0, scale=sd).rvs(shape)
        self.w_hist.append(np.mean(init_w[0:10,:], axis=1))
        return init_w

    
    def generate_mask(self):
        """
        return np.array
            a mask of inverted dropout
        """
        size = self.unit_num
        if self.fix_seed:
            np.random.seed(seed=1)
        inverted_mask = binomial(n=1, p=self.keep_p, size=size) * (1/self.keep_p)
        self.inverted_drop_mask = inverted_mask.reshape(1, size)
        return self.inverted_drop_mask
    
    def apply_gradient(self, nn):
        """
        applying the new updated weights as the current weights
        ----------
        nn: object of Neural Network class
        """
        grad = self.diff_w
        lr = nn.learning_rate
        
        # Adjust grad
        if nn.w_decay_lambda is not None:
            if self.with_bias: # weight decay does not effect on bias
                grad[:, :-2] -= nn.w_decay_lambda * self.w[:, :-2]
            else:
                grad -= nn.w_decay_lambda * self.w
        
        # Apply grad into weight
        if nn.momentum is not None:
            grad = nn.momentum * self.prev_diff_w + lr * grad
            self.w += grad
            self.prev_diff_w = grad
        else:
            self.w += lr * grad
        
    @classmethod
    def merge_with_bias(cls, org_):
        """
        return np.array
            it returns org_ merged with a bias vector or matrix
        ----------
        org_: numpy.array
            numpy array which will be merged a bias vector or matrix
        """
        return np.concatenate((org_, np.ones((org_.shape[0], 1))), axis=1)
```

## Neural Network class
It represents a neural network

```python
class NeuralNetwork:
    """
    Representing a neural network
    """
    def __init__(self, structures, f_cost, learning_rate=0.01, with_bias=True,
                 w_decay_lambda=None, momentum=None, lr_updator=None, fix_seed=False):
        """
        structures: dict
            a list of dict consisting of "type": Enum of LType, "unit_num": (int) the number of units, "f_act": activation function, 
        f_cost: function
            it is used for calculating cost
        learning_rate: float
            learning rate from 0 to 1 used for updating weights
        with_bias: boolean
            true if introducing bias in the input and hidden layers
        w_decay_lambda: float
            coefficient from 0 to 1 multiplying into the weight while updating weight
        momentum: float
            coefficient from 0 to 1 multiplying into the moment while updating weight
        lr_updator: function
            the specified function should take the current epoch number as an argument
            and return the corresponding learning rate.
        fix_seed: boolean
            true if removing randomness; for debugging purpose
        """
        self.f_cost = f_cost # cost function
        self.learning_rate = learning_rate
        self.with_bias = with_bias
        self.fix_seed = fix_seed
        self.w_decay_lambda = w_decay_lambda
        self.momentum = momentum
        self.layers = self.__init_layers(structures)
        self.lr_updator = lr_updator
        self.early_stop = None # it is specified when starting fit
        self.last_epoch = 0
        self.train_accuracies = [] # train accuracy history
        self.test_accuracies = [] # test/valid accuracy history
        self.train_losses= [] # train cost history
        self.test_losses = [] # test/train cost history
        self.test_recall = []
        self.test_precision = []
        self.test_output_mean = []
        
    def __init_layers(self, structures):
        """
        return a list of initialized Layer class
        ----------
        structures: a list of dict
            please see Layer.__init__ for the further information about a content of dict
        """
        ealier_unit_num = -1
        new_layers = []
        for i in range(len(structures)):
            new_layers.append(Layer(self, structures[i], ealier_unit_num, self.with_bias, self.fix_seed))
            ealier_unit_num = new_layers[i].unit_num

        return new_layers
    
    def predict(self, X_test):
        """
        return predicted labels
        ----------
        X_test: numpy.array
            it should be features for the prediction
        """
        return self.__passing_forward(np.array(X_test, ndmin=2), is_training=False)
            
    def show_confusion_matrix(self, X_test, y_test, pred_=None, mapper=None):
        """
        return a dataframe showing the confusion matrix between predictions and true labels
        ----------
        X_test: numpy.array
            features to get predicted labels
        y_test: numpy.array
            true labels
        pred: numpy.array
            predicted labels
        """ 
        size = self.layers[-1].unit_num - 1
        cm = np.zeros((size, size), int)
        pred = self.predict(X_test) if pred_ is None else pred_
        pred_max = pred.argmax(axis=1)
        label_max = np.array(y_test, ndmin=2).argmax(axis=1)
        for row_, col_ in zip(pred_max, label_max):
            cm[row_, col_] += 1
        cm = pd.DataFrame(cm)
        cm.index.name = "Prediction"
        cm.columns.name = "True label"
        
        # return matrix for measuring performance
        if pred_ is not None:
            return cm
        # return matrix for showing confusion matrix
        else:
            return self.__get_cm_with_recall_prec(cm, mapper=mapper)
        
    
    def __calc_loss(self, pred, labels):
        """
        returns float
            it returns a cost based on the cost function
        ----------
        pred: np.array
            features to make predictions
        labels: np.array
            true labels
        """ 
        loss = self.f_cost(pred, labels)
        # calculate extra cost when using weight decay
        if self.w_decay_lambda is not None:
            for layer in self.layers:
                if layer.type in [LType.HID, LType.OUT]:
                    if self.with_bias: # bias is not penalized
                        loss += 0.5 * self.w_decay_lambda * np.sum(layer.w[:, :-2] ** 2)
                    else:
                        loss += 0.5 * self.w_decay_lambda * np.sum(layer.w ** 2)
            
        return loss
                
    def __get_performance(self, feats, labels, is_valid=False):
        """
        returns an accuracy
        ----------
        feats: np.ndarray
            features to make predictions
        labels: np.array
            true labels
        """ 
        recall, prec = [], []
        pred = self.predict(feats)
        cm = self.show_confusion_matrix(feats, labels, pred)
        acc = np.sum(np.diagonal(cm))/np.sum(cm.values)
        loss = self.__calc_loss(pred, labels)
        if is_valid:
            self.__calc_output_mean(pred, labels)
            cm = self.__get_cm_with_recall_prec(cm)
            recall, prec = self.__calc_recall_prec(cm, labels.shape[1])
        return acc, loss, recall, prec
    
    def __calc_recall_prec(self, cm, class_num):
        """
        returns np.array
            calculated recall and proc
        ----------
        cm: pd.DataFrame
            confusion matrix
        class_num: int
            the number of classes
        """ 
        recall = cm.loc["Recall", :].iloc[0:class_num].values
        prec = cm.loc[:, "Precision"].iloc[0:class_num].values
        return recall, prec
    
    def __get_cm_with_recall_prec(self, cm, mapper=None):
        """
        returns pd.Dataframe
            add Recall and Precision into confusion matrix given
        ----------
        cm: pd.DataFrame
            confusion matrix
        mapper: dict
            convert idx and column names corresponding key in dict into values
        """ 
        diag = np.diagonal(cm)
        pres = np.round(diag / np.sum(cm, axis=1), 2).fillna(0)
        pres.name = "Precision"
        cm = pd.concat([cm, pres], axis=1)
        rec = np.round(diag / np.sum(cm, axis=0)[:-1], 2).fillna(0)
        cm.loc["Recall"] = np.concatenate((rec.values, np.array([np.NaN])))
        cm.columns.name = "True label"
        if mapper is not None:
            cm = cm.rename(mapper=mapper, axis=0).rename(mapper=mapper, axis=1)
        return cm
    
    def __calc_output_mean(self, pred, labels):
        """
        Store output mean into the object(self.test_output_mean)
        ----------
        pred: np.array
            predicted values
        labels: np.array
            true values
        """ 
        num_class = labels.shape[1]
        op_means = []
        for i in range(num_class):
            # extract idx where the correct label is zero
            idx_ = np.where(labels[:, i] == 1)
            # store predicted values corresponding the idx
            op_means.append(np.mean(pred[idx_], axis=0))
        self.test_output_mean.append(np.vstack(op_means))
    
    def __epoch_info_collection(self, X_train, y_train, X_test, y_test):
        """
        store accuracy and loss into instance variable of Neural Network about the current epoch
        ----------
        X_train: np.ndarray
            training features to get predicted labels
        y_train: np.array
            true labels corresponding to X_train
        X_test: np.array
            validating features to get predicted labels
        y_test: np.array
            true labels corresponding to X_test
        """ 
        test_acc, test_loss, recall, prec = self.__get_performance(X_test, y_test, is_valid=True)
        train_acc, train_loss, _, _ = self.__get_performance(X_train, y_train)
        self.test_accuracies.append(test_acc)
        self.test_losses.append(test_loss)
        self.train_accuracies.append(train_acc)
        self.train_losses.append(train_loss)
        self.test_recall.append(recall)
        self.test_precision.append(prec)
        
        
    def __minibatch_info_collection(self):
        for layer in self.layers:
            if layer.type in [LType.HID, LType.OUT]:
                layer.w_hist.append(np.mean(layer.w[0:10,:], axis=1))
        
    def fit_with_info_collection(self, X_train, y_train, X_test, y_test, batchsize=32, epoch=1, early_stop=None):
        """
        call NeuralNetwork.fit with validation data for collecting information of accuracy and loss.
        Please refer NeuralNetwork.fit to know the arguments.
        """         
        self.fit(X_train, y_train, X_test=X_test, y_test=y_test, batchsize=batchsize, epoch=epoch, early_stop=early_stop)
    
    def fit(self, X_train, y_train, X_test=None, y_test=None, batchsize=32, epoch=1, early_stop=None):
        """
        train neural network model
        ----------
        X_train: np.array
            features for training
        y_train: np.array
            labels for training
        X_test: np.array
            features for getting test accuracy and loss
        y_test: np.array
            labels for getting test accuracy and loss
        batchsize: int
            the size of minibatch
        early_stop: EarlyStopping class
            it controls when stopping fitting
        """ 
        self.early_stop = early_stop
        time_total = 0
        iter_per_epoch = max(np.floor(len(y_train) / batchsize), 1).astype(np.int32) # take max for in case that y_train < batchsize
        X_train = np.array(X_train, ndmin=2)
        y_train = np.array(y_train, ndmin=2)
        print(f"Start training ...")
        for epc in range(self.last_epoch + 1, epoch + self.last_epoch + 1):
            st = time.time()
            self.learning_rate = self.lr_updator(epc) if self.lr_updator is not None else self.learning_rate

            # training with 1 mini-batch
            for i in range(iter_per_epoch):
                if self.fix_seed: # debug purpose
                    np.random.seed(seed=1)
                idx = np.random.choice(len(y_train), batchsize) # get different idx every single mini-batch
                mini_inputs = X_train[idx]
                mini_labels = y_train[idx]
                
                self.__passing_forward(mini_inputs)
                self.__calc_gradients(mini_labels)
                self.__apply_gradients()
                self.__minibatch_info_collection()

            # collect information of acc and loss
            if (X_test is not None) & (y_test is not None):
                self.__epoch_info_collection(X_train, y_train, X_test, y_test)
                train_loss = f", train_loss: {np.round(self.train_losses[-1], 4)}"

            # calculate time passed per 1 epoch
            time_epoch = time.time() - st
            time_total += time_epoch
            self.last_epoch = epc
            print(f"[Finish epoch {epc}] total process time {np.round(time_total, 1)} sec{train_loss}")
            
            # control early stopping
            if self.early_stop is not None:
                self.early_stop.patient_counter(self)
                if self.early_stop.judge_stop(self):
                    break
                
    def __passing_forward(self, input_vector, is_training=True):
        """
        return np.array
            It performs passing forward, stores computational results into instance variable of Layer,
            returns the output of the network.
        ----------
        input_vector: np.array
            input of the neural network
        is_training: boolean
            True for the training, False for the prediction
        """ 
        layers = self.layers
        for i in range(len(layers)):
            if layers[i].type == LType.INP:
                if layers[i].with_bias:
                    layers[i].x = layers[i].merge_with_bias(input_vector)
                else:
                    layers[i].x = input_vector
                    
                layers[i].output = layers[i].x

                if (layers[i].keep_p is not None) & is_training:
                    layers[i].output *= layers[i].generate_mask()
                    
            elif layers[i].type == LType.HID:
                layers[i].x = layers[i-1].output
                layers[i].wx = np.dot(layers[i].w, layers[i].x.T).T

                if layers[i].with_bias:
                    layers[i].output = layers[i].merge_with_bias(layers[i].f_act(layers[i].wx))
                else:
                    layers[i].output = layers[i].f_act(layers[i].wx)
                    
                if (layers[i].keep_p is not None) & is_training:
                    layers[i].output *= layers[i].generate_mask()
                    
            else: # LType.OUT
                layers[i].x = layers[i-1].output
                layers[i].wx = np.dot(layers[i].w, layers[i].x.T).T
                layers[i].output = layers[i].f_act(layers[i].wx)
        
        return layers[-1].output # return the output of the last layer

   
    def __calc_gradients(self, target_vector):
        """
        It performs backpropagation, calculate gradients and store them into layers.
        ----------
        target_vector: np.array
            true labels corresponding to the input specified in passing forward
        """ 
        # reverse layer order
        layers = self.layers[::-1]
        for i in range(len(layers)):
            # backprop for OUTPUT layer
            if layers[i].type == LType.OUT:
                # calc partial derivatives denoted as pd_***
                pd_err = differentiate_cost_f(self.f_cost, layers[i].f_act,
                                              target_vector, layers[i].output)
                pd_op = differentiate_act_f(layers[i].f_act, layers[i].output) # 
                delta = pd_err * pd_op
                
                # keep temporal results for the subsequent processing
                layers[i].diff_w = np.dot(delta.T, layers[i].x)
                layers[i].delta = delta
            # backprop for HIDDEN layer
            elif layers[i].type == LType.HID:
                # calc partial derivatives denoted as pd_***
                pd_err = np.dot(layers[i-1].delta, layers[i-1].w)
                pd_op = differentiate_act_f(layers[i].f_act, layers[i].output)
                delta = (pd_err * pd_op)[:, :-1] if layers[i].with_bias else (pd_err * pd_op) # [:, :-1] remove bias part
                
                # keep temporal results for the subsequent processing
                layers[i].diff_w = np.dot(delta.T, layers[i+1].output)
                layers[i].delta = delta

    def __apply_gradients(self):
        """ update original weights by using calculated gradients """
        for layer in self.layers:
            if layer.type in [LType.HID, LType.OUT]:
                layer.apply_gradient(self)
```

## Early Stopping class

```python
class EarlyStopping:
    """
    Controlling the early stopping.
    """
    def __init__(self, monitor="test_losses", min_delta=0, patience=0):
        """
        monitor: property name in NeuralNetwork class
            this variable is used to measure the performance of training and to judge delta and patience
        min_delta: float or int
            increment patience if the performance of the monitored property is lower than min_delta
        patience: int
            count patience if a monitored property shows a lower performance than the previous epoch/min_delta
            stop fitting if the count exceeds the specified patience here
        """ 
        self.monitor = monitor
        self.min_delta = min_delta
        self.pat_limit = patience
        self.pat_cnt = 0

    def patient_counter(self, nn):
        """
        It counts the patient if a monitored value shows a lower performance than the previous epoch
        ----------
        nn: object of NeuralNetwork class
        """ 
        monitored_hist = self.__get_monitored_hist(nn)
        
        # skip count in the 1st epoch
        if len(monitored_hist) > 1:
            # multiply -1 if monitored value is accuracy, 1 if it is loss
            cur_val  = monitored_hist[-1] * self.__sign()
            prev_val = monitored_hist[-2] * self.__sign()
            
            # judge whether monitored value corresponds to the criteria of patient
            if (cur_val >= prev_val) | ((prev_val - cur_val) <= self.min_delta):
                self.pat_cnt += 1
                print(f"[Early stop cnt] monitoring: {self.monitor}, cnt:{self.pat_cnt}, prev: {prev_val}, cur: {cur_val}")
            else:
                self.pat_cnt = 0

    def judge_stop(self, nn):
        """
        return Boolean
            It judges whether fitting should be stopped or not
        ----------
        nn: object of NeuralNetwork class
        """ 
        if self.pat_cnt > self.pat_limit:
            hist = self.__get_monitored_hist(nn)
            hist_with_patient = hist[-1 * self.pat_cnt:]
            print(f"[Early stopping] Recent transition of {self.monitor}: {hist_with_patient}")
            return True
        else:
            return False
        
    def __get_monitored_hist(self, nn):
        """
        return list
            It returns a monitored value in an object of NeuralNetwork
        ----------
        nn: object of NeuralNetwork class
        """ 
        return getattr(nn, self.monitor)
    
    def __sign(self):
        """
        return int
            It adjusts sign of accuracy and cost.
            Return 1 if loss such that a larger value is a worse value.
            Return -1 if not loss like accuracy such that a larger value is a better value.
        """ 
        if "loss" in self.monitor:
            return 1
        else: # expect "accuracies"
            return -1
```
